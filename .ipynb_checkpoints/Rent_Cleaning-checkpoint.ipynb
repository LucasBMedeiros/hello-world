# Data Cleaning

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('df1_unclean.csv')
```

```python
df1 = df1[['NOME', 'VALOR', 'DETALHES', 'ENDEREÇO']]
```

```python
df1
```

```python
df1.info()
```

```python
df1.describe()
```

```python
df1['ENDEREÇO'].unique()
```

```python
import re
```

```python
f_quarto = lambda x: re.findall('[0-9]+ quartos?', str(x))
f_area = lambda x: re.findall('[0-9]+m²', str(x))
f_condominio = lambda x: re.findall('Condomínio: R\$\ [0-9]+', str(x))
f_vaga = lambda x: re.findall('[0-9]+ vagas?', str(x))
f_valor = lambda x: x.split()[-1] if x.split() else ''
```

```python
df2 = pd.DataFrame()
df2['N° QUARTOS'] = list(map(f_quarto, df1['DETALHES']))
df2['ÁREA (m²)'] = list(map(f_area, df1['DETALHES']))
df2['CONDOMINIO'] = list(map(f_condominio, df1['DETALHES']))
df2['VAGAS'] = list(map(f_vaga, df1['DETALHES']))
df2['VALOR'] = list(map(f_valor, df1['VALOR'].astype(str)))
df2
```

```python
f1_quarto = lambda x: re.findall('[0-9]+', str(x))
f1_area = lambda x: re.findall('[0-9]+', str(x))
f1_condominio = lambda x: re.findall('[0-9]+', str(x))
f1_vaga = lambda x: re.findall('[0-9]+', str(x))

df3 = pd.DataFrame()
df3['N° QUARTOS'] = list(map(f1_quarto, df2['N° QUARTOS']))
df3['ÁREA (m²)'] = list(map(f1_area, df2['ÁREA (m²)']))
df3['CONDOMINIO'] = list(map(f1_condominio, df2['CONDOMINIO']))
df3['VAGAS'] = list(map(f1_vaga, df2['VAGAS']))
df3['VALOR'] = df2['VALOR']
df3[['NOME', 'ENDEREÇO']] = df1[['NOME', 'ENDEREÇO']]
```

```python
df3
```

```python
df4 = pd.DataFrame()
df4 = df3[['NOME', 'ENDEREÇO', 'N° QUARTOS', 'ÁREA (m²)', 'CONDOMINIO', 'VAGAS', 'VALOR']]
df4
```

# N° Quartos

```python
n = df4['N° QUARTOS'].astype(str)
def content_from_lists(n):
    lst = []
    for i in n:
        if len(i)>3:
            lst.append(i[-3])
        else:
            lst.append('') # valores vazios são interpretados como kitnets e inse-re-se 1 quarto
    lst = pd.to_numeric(lst)
    return lst
    print('The result is a list "lst": ', lst)
    
content_from_lists(n)
```

```python
a = pd.Series(content_from_lists(n), dtype=int)
print(a)
```

```python
df4['N° QUARTOS'] = a
```

# Valor

```python
ltr = []
for i in df4['VALOR'].to_list():
    if i == 'nan':
        ltr.append(0)
    else:
        ltr.append(i.replace('.', ''))
ltr = pd.to_numeric(ltr)
```
